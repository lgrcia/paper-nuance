```python
import pandas as pd

df = pd.read_csv(
    "/Users/lgrcia/code/studies/paper-nuance/workflows/real_injection_recovery/data/29923559/search_results.csv"
)
```

```python
tls, nuance, radii, periods, tau, delta = (
    df["tls_found"].values,
    df["nuance_found"].values,
    df["planet_radius"].values,
    df["planet_period"].values,
    df["tau"].values,
    df["delta"].values,
)
```

```python
import matplotlib.pyplot as plt
import numpy as np


# setting up ranges
taus_range = np.min(tau), np.max(tau)
deltas_range = np.min(delta), np.max(delta)
periods_range = np.min(periods), np.max(periods)
radii_range = np.min(radii), np.max(radii)
extent = (*periods_range, *radii_range)

# figure
fig = plt.figure(None, (7, 4))

ax = plt.subplot(121)
plt.imshow(nuance.reshape(20, 20), extent=extent, origin="lower", aspect="auto")
plt.xlabel("Period (days)")
plt.ylabel("Radius (R$_\oplus$)")
# secondary axes
# secax = ax.secondary_xaxis("top")
# secax.set_xticks(periods_range, [f"{t:.1f}" for t in taus_range])
# secax.set_xlabel(r"$\tau$")
ax.set_title("nuance", loc="left")

ax = plt.subplot(122)
plt.imshow(tls.reshape(20, 20), extent=extent, origin="lower", aspect="auto")
plt.xlabel("Period (days)")
# secondary axes
# secax = ax.secondary_xaxis("top")
# secax.set_xticks(periods_range, [f"{t:.1f}" for t in taus_range])
# secax.set_xlabel(r"$\tau$")
# secax = ax.secondary_yaxis("right")
# ax.yaxis.set_visible(False)
# secax.set_yticks(radii_range, [f"{t:.1f}" for t in deltas_range])
# secax.set_ylabel(r"$\delta$")
ax.set_title("w√µtan + TLS", loc="left")

fig.suptitle(
    r"TIC 354534653"
    + "\n"
    + rf"$\tau \in [{taus_range[0]:.1f}, {taus_range[1]:.1f}]$"
    + " and "
    + rf"$\delta \in [{deltas_range[0]:.1f}, {deltas_range[1]:.1f}]$"
)

plt.tight_layout()
```
