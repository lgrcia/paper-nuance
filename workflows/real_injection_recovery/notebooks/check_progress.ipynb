```python
from datetime import datetime

# read the log file

datetimes = []
# iterate over files

for f in [
    "/Users/lgrcia/code/studies/paper-nuance/workflows/real_injection_recovery/.snakemake/log/2023-04-29T153929.755430.snakemake.log",
    "/Users/lgrcia/code/studies/paper-nuance/workflows/real_injection_recovery/.snakemake/log/2023-04-30T142310.313877.snakemake.log",
]:
    with open(f, "r") as f:
        log_lines = f.readlines()

    for i in range(len(log_lines)):
        if "rule nuance" in log_lines[i] or "rule tls" in log_lines[i]:
            date = " ".join(log_lines[i - 1].split()[1:5]).replace("]", "")
            datetimes.append(datetime.strptime(date, "%b %d %H:%M:%S %Y"))
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(None, (10, 5))
plt.plot(datetimes, np.arange(len(datetimes)))
plt.tight_layout()
```
